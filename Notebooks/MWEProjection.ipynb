# Objective

Let $\Omega$ be an on empty set and $S \subset \Omega$ such that $|S| > 2$. Let
$$\mathcal{A}(S) = \Bigg\{(W, b) \in \mathbb{R}^{\Omega} \times \mathbb{R} ~|~ \sum_{k \in \Omega \setminus S, w_k \geq 0}w_k \leq b \leq \min_{s \in S}w_i + \sum_{k \in \Omega \setminus S, w_k < 0}w_k\Bigg\}$$

Then $\mathcal{A}(S)$ is a convex set as the intersection of half-planes. Moreover, it is closed. We want to compute the projection onto $\mathcal{A}(S)$. Let $(\hat W, \hat b) \in (\mathbb{R}_+^{\Omega} \times \mathbb{R}) \setminus \mathcal A(S)$ be a set of positive weights and bias.
$$
\begin{align}
&\min_{(W, b) \in \mathbb{R}^{\Omega} \times \mathbb{R}}\frac{1}{2}||W - \hat W||^2_2 + \frac{1}{2}(b - \hat b)^2 = \frac{1}{2}\sum_{k \in \Omega}(w_k - \hat w_k)^2 + \frac{1}{2}(b - \hat b)^2 \\
&\text{s.t.}
\begin{cases}
\sum_{i \in \Omega \setminus S, w_i \geq 0}w_i - b \leq 0\\
\forall s \in S , b -w_s - \sum_{i \in \Omega \setminus S, w_i < 0}w_i \leq 0
\end{cases} \Leftrightarrow 
\begin{cases}
\sum_{i \in \Omega \setminus S}\max(w_i, 0) - b \leq 0\\
\forall s \in S , b -w_s + \sum_{i \in \Omega \setminus S}\max(-w_i, 0) \leq 0
\end{cases}
\end{align}
$$

We define
$$\begin{align}
\mathbb K^+ &= \{k \in \Omega \setminus S ~|~ w_k \geq 0\} \\
\mathbb K^- &= \{k \in \Omega \setminus S ~|~ w_k < 0\} \\
\end{align}
$$
and we have $\Omega \setminus S = \mathbb K^+ \cup \mathbb K^-$.



# Lagrangian

To solve this constrained optimization problem, we define the Lagrangian function:
$$\mathcal{L}(W, b, \lambda) = \frac{1}{2}\sum_{k \in \Omega}(w_k - \hat w_k)^2 + \frac{1}{2}(b - \hat b)^2 + \lambda_0\Big(\sum_{i \in \mathbb K^+}w_i - b\Big) + \sum_{s \in S}{\lambda_s\Big(b - w_s - \sum_{k \in \mathbb K^-}w_k\Big)}
$$

We find that $(\mathbb{1}_S, \frac{1}{2}) \in \text{int}(\mathcal{A}(S))$, with $\mathbb{1}_S$ the indicator function of $S$. The objective function and constraints are convex. Therefore, Slater conditions hold, and there exist $(W, b, \lambda, \lambda_0) \in \mathbb{R}_+^{\Omega} \times \mathbb R \times \mathbb{R}_+^S \times \mathbb R_+$ such that

$$\begin{align}
\forall k \in \mathbb K^+, \frac{\partial \mathcal{L}}{\partial w_k} &= w_k - \hat w_k + \lambda_0 = 0  \\
\forall k \in \mathbb K^-, \frac{\partial \mathcal{L}}{\partial w_k} &= w_k - \hat w_k - \sum_{s \in S}\lambda_s = 0\\
\forall s \in S, \frac{\partial \mathcal{L}}{\partial w_s} &= w_s - \hat w_s - \lambda_s = 0\\
\frac{\partial \mathcal{L}}{\partial b} &= b - \hat b - \lambda_0 + \sum_{s \in S}\lambda_s = 0\\
\forall s \in S, \lambda_s\Big(b - w_s - \sum_{k \in \mathbb K^-}w_k \Big) &= 0 \\
\lambda_0 \Big(\sum_{k \in \mathbb K^+}w_k -b\Big) &= 0
\end{align}
$$

Using equation (2), we find that $\mathbb K^- \subset \{k \in \Omega \setminus S ~|~ \hat w_k < 0\} = \emptyset$. Therefore, the equations become

$$\begin{align}
\forall k \in \Omega \setminus S, \frac{\partial \mathcal{L}}{\partial w_k} &= w_k - \hat w_k + \lambda_0 = 0  \\
\forall s \in S, \frac{\partial \mathcal{L}}{\partial w_s} &= w_s - \hat w_s - \lambda_s = 0\\
\frac{\partial \mathcal{L}}{\partial b} &= b - \hat b - \lambda_0 + \sum_{s \in S}\lambda_s = 0\\
\forall s \in S, \lambda_s\Big(b - w_s \Big) &= 0 \\
\lambda_0 \Big(\sum_{k \in \Omega \setminus S}w_k -b\Big) &= 0
\end{align}
$$



# Resolution 
We define $$T = \{t \in S ~|~ w_t = b\}$$
Therefore using equation (4), $$\forall j \in S \setminus T, \lambda_j = 0$$
Then using equation (2), $\forall t \in T, \hat w_t \leq w_t = b$. Reciprocally, let $s \in S$ such that $\hat w_s \leq b$. If $s \notin T$, then $\lambda_s = 0$ and $b \geq \hat w_s = w_s > b$, which is absurd. Therefore,
$$T = \{t \in S ~|~ \hat w_t \leq b\}
$$


## Case 1: $|T| = 0$ and $\sum_{k \in \Omega \setminus S}w_k \ne b$

Then $\forall s \in S, \lambda_s = \lambda_0 = 0$, therefore $(W, b) = (\hat W, \hat b)$. This is not possible as $(\hat W, \hat b) \notin \mathcal A(S)$ by hypothesis.



## Case 2: $|T| > 0$ and $\sum_{k \in \Omega \setminus S}w_k \ne b$ (Right)

Then $\lambda_0 = 0$. Then $\sum(2)$ and $(3)$ gives:
$$\sum_{t \in T}(b - \hat w_t) = \hat b - b \Leftrightarrow b = \frac{1}{|T| + 1}\Big(\sum_{t \in T} \hat w_t + \hat b\Big)$$

Finally
$$\begin{align}
b &= \frac{1}{|T| + 1}\Big(\sum_{t \in T} \hat w_t + \hat b\Big)\\
\forall k \in \Omega \setminus T, w_k &= \hat w_k \\
\forall t \in T, w_t &= b = \frac{1}{|T| + 1}\Big(\sum_{t \in T} \hat w_t + \hat b\Big)\\
\forall t \in T, \lambda_t &= \frac{1}{|T| + 1}\Big(\sum_{t \in T} \hat w_t + \hat b\Big) - \hat w_t
\end{align}
$$


## Case 3: $|T| = 0$ and $\sum_{k \in \Omega \setminus S}w_k = b$ (Left)

Then $\forall s \in S, \lambda_s = 0$. Then $\forall k \in \Omega \setminus S, \hat w_k - w_k = b - \hat b$. We sum on all $k$: 
$$\sum_{k \in \Omega \setminus S}\hat w_k - b = |\Omega \setminus S|(b - \hat b) \Leftrightarrow b = \frac{1}{|\Omega \setminus S| + 1} \Big(\sum_{k \in \Omega \setminus S}\hat w_k + |\Omega \setminus S| \hat b\Big)$$

Finally
$$\begin{align}
b &= \frac{1}{|\Omega \setminus S| + 1} \Big(\sum_{k \in \Omega \setminus S}\hat w_k + |\Omega \setminus S| \hat b\Big) \\
\forall s \in S, w_s &= \hat w_s\\
\forall k \in \Omega \setminus S, w_k &= \hat w_k - \frac{1}{|\Omega \setminus S| + 1} \Big(\sum_{k \in \Omega \setminus S}\hat w_k - \hat b\Big)\\
\lambda_0 &= \frac{1}{|\Omega \setminus S| + 1} \Big(\sum_{k \in \Omega \setminus S}\hat w_k - \hat b\Big)
\end{align}$$

$\renewcommand{\card}[1]{|#1|}$


## Case 4: $|T| > 0$ and $\sum_{k \in \Omega \setminus S}w_k = b$ (Left Right)

The set of equation becomes
$$\begin{align}
\forall i \in \Omega \setminus S, w_i + \lambda_0 &= \hat w_i\\
\forall t \in T, b - \lambda_t &= \hat w_t\\
b - \lambda_0 + \sum_{t \in T}\lambda_t &= \hat b\\
\sum_{i \in\Omega\setminus S}w_i - b &= 0
\end{align}
$$


Let$$\begin{align}
a &= |T|\\
c &= |\Omega\setminus S|
\end{align}$$
Let $I_n$ be the identity matrix of size $n$.  Let $\mathbf1_n$ be the column vector of only 1 of size $n$. Let $\mathbf0_{n, n}$ be the matrix of only $0$ of size $n$. Then we define the matrix system:
$$\begin{align}
A &= \begin{pmatrix}
I_{c} & \mathbf1_{c} & \mathbf0_{c\times a } & \mathbf0_{c}\\
\mathbf0_{a \times c} & \mathbf0_{a} & -I_{a} & \mathbf1_{a}\\
\mathbf0_{c}^T & -1 & \mathbf1_{a}^T & -1\\
\mathbf{1}_{c}^T & 0 & \mathbf0_{a}^T & -1
\end{pmatrix} \in \R^{(c + a + 2) \times (c + a + 2)}\\
X &= \begin{pmatrix}
(w_i)_{i \in \Omega\setminus S}\\
\lambda_0\\
(\lambda_t)_{t \in T}\\
b
\end{pmatrix} \in \R^{c + a + 2}\\
Y &= \begin{pmatrix}
(\hat w_i)_{i \in \Omega\setminus S}\\
(\hat w_t)_{t \in T}\\
\hat b\\
0
\end{pmatrix} \in \R^{c + a + 2}
\end{align}
$$
Let$$\begin{align}
\tilde a &= \frac{|T| - 1}{1 + \card{\Omega \setminus S}(|T| - 1)}\\
\tilde c &= \frac{\card{\Omega\setminus S}}{{1 + \card{\Omega \setminus S}(|T| - 1)}}\\
\tilde d &= \frac{1}{{1 + \card{\Omega \setminus S}(|T| - 1)}}
\end{align}$$
Then the inverse matrix is 
$$A^{-1} = \begin{pmatrix}
I_{c} - \tilde a\mathbf\cdot1_{c \times c} & \tilde d \cdot \mathbf1_{c\times a} & \tilde d\cdot\mathbf1_c &\tilde a\cdot\mathbf1_{c}\\
\tilde a\cdot\mathbf1_c^T & -\tilde d \cdot\mathbf1_{a}^T & -\tilde d & -\tilde a\\
\tilde d \cdot \mathbf1_{a \times c} & \tilde c\cdot\mathbf1_{a\times a}-I_a & \tilde c\cdot\mathbf{1}_a & -\tilde d \cdot\mathbf1_a\\
\tilde d\cdot \mathbf1_{c}^T & \tilde c\cdot\mathbf1_a^T & \tilde c & -\tilde d
\end{pmatrix} \in \R^{(c + a + 2) \times (c + a + 2)}
$$

Then:
$$\begin{align}
\forall i \in \Omega \setminus S, w_i &= \hat w_i - \tilde a\sum_{k \in \Omega \setminus S}\hat w_k + \tilde d \Big(\sum_{t \in T}\hat w_t + \hat b\Big) = \hat w_i + \frac{1}{1 + \card{\Omega\setminus S}(\card T - 1)}\Bigg(\sum_{t \in T}\hat w_t + \hat b - (\card T - 1)\sum_{k \in  \Omega \setminus S}\hat w_k\Bigg)\\
\forall t \in T, \lambda_t &= \tilde d \sum_{k \in \Omega \setminus S}\hat w_k - \hat w_t + \tilde c\Big(\sum_{j \in T}\hat w_j + \hat b\Big) = -\hat w_t + \frac{1}{1 + \card{\Omega\setminus S}(\card T - 1)}\Bigg(\card{\Omega \setminus S}\Big(\sum_{j \in T}\hat w_j + \hat b\Big) + \sum_{k \in \Omega \setminus S}\hat w_k\Bigg)\\
\lambda_0 &= \tilde a \sum_{k \in \Omega \setminus S}\hat w_k - \tilde d \Big(\sum_{t \in T}\hat w_t + \hat b\Big) = \frac{1}{1 + \card{\Omega\setminus S}(\card T - 1)}\Bigg(-\Big(\sum_{t \in T}\hat w_t + \hat b\Big) + \Big(\card{T} - 1\Big)\sum_{k \in \Omega \setminus S}\hat w_k\Bigg)\\
b &= \tilde d\sum_{k \in \Omega \setminus S}\hat w_k + \tilde c \Big(\sum_{t \in T}\tilde w_t + \hat b\Big) = \frac{1}{1 + \card{\Omega\setminus S}(\card T - 1)}\Bigg(\card{\Omega \setminus S}\Big(\sum_{t \in T}\hat w_t + \hat b\Big) + \sum_{k \in \Omega \setminus S}\hat w_k\Bigg)
\end{align}
$$


# Aggregation

For each case, we gather union the possible couple $\{(W, b)\}$ for case 3 and $\Bigg\{\Big(W(T), b(T)\Big) \Big| T = \{t \in S | w_t \leq w_k\}, k \in S \Bigg\}$ for cases 2 and 4. By Slater conditions, we know that the optimal value is inside this set. We remind that as $\hat W \geq 0$, then the best projection $W^* \geq 0$. We gather all couples where the constraints are met, of both primal and dual:
$$\begin{align}
\sum_{k \in \Omega \setminus S}w_k \leq b \leq \min_{s \in S}w_s \\
\lambda_0 \geq 0 \\
\forall s \in S, \lambda_s \geq 0
\end{align}
$$
 We compute the loss for each candidate and take the lowest loss.

```python
def system4(Sbar, T):
    a = T
    c = Sbar

    A = np.zeros((Sbar + T + 2, Sbar + T + 2))
    
    A[:c, :c] = np.eye(c)
    A[:c, c] = 1
    
    A[c:c+a, c+1:c+1+a] = -np.eye(a)
    A[c:c+a, -1] = 1

    A[c+a, c] = -1
    A[c+a, c+1:c+1+a] = 1
    A[c+a, -1] = 1

    A[-1, :c] = 1
    A[-1, -1] = -1
    return A

def system4inv(Sbar, T):
    denom = 1 / (1 + Sbar * (T + 1))
    d = denom
    a = denom * (T + 1)
    c = Sbar * denom

    Ainv = np.zeros((Sbar + T + 2, Sbar + T + 2))
    Ainv[:Sbar, :Sbar] = np.eye(Sbar) - a
    Ainv[:Sbar, Sbar: Sbar + T + 1] = d
    Ainv[:Sbar, -1] = a

    Ainv[Sbar, :Sbar] = a
    Ainv[Sbar, Sbar: Sbar + T + 1] = -d
    Ainv[Sbar, -1] = -a

    Ainv[Sbar + 1: Sbar + T + 1, :Sbar] = d
    Ainv[Sbar + 1: Sbar + T + 1, Sbar: Sbar + T] = c - np.eye(T)
    Ainv[Sbar + 1: Sbar + T + 1, Sbar + T] = c
    Ainv[Sbar + 1: Sbar + T + 1, -1] = -d

    Ainv[-1, :Sbar] = d
    Ainv[-1, Sbar:-1] = c
    Ainv[-1, -1] = -d

    return Ainv
```

# Projection

```python
class Projection:
    
    def __init__(self, Wini: np.ndarray = None, bini: np.ndarray = None, S: np.ndarray = None, T: np.ndarray= None, epsilon: float = 1e-4):
        self.Wini = Wini
        self.bini = bini
        self.S = S
        self.T = T
        self.epsilon = epsilon

        self.b = None
        self.lambda0: float = None
        self.lambdat: np.ndarray = None
        self.W = None
        self.loss = None
        self.summary = None

    @staticmethod
    def left_cond(Wini: np.ndarray = None, bini: np.ndarray = None, S: np.ndarray = None, epsilon: float = 1e-4):
        return left_cond(Wini=Wini, bini=bini, S=S, epsilon=epsilon)
        
    @staticmethod
    def right_cond(Wini: np.ndarray = None, bini: np.ndarray = None, S: np.ndarray = None, epsilon: float = 1e-4):
        return right_cond(Wini=Wini, bini=bini, S=S, epsilon=epsilon)

    @staticmethod
    def left_margin(Wini: np.ndarray = None, bini: np.ndarray = None, S: np.ndarray = None):
        lb = Wini[~S].sum()
        return bini - lb

    @staticmethod
    def right_margin(Wini: np.ndarray = None, bini: np.ndarray = None, S: np.ndarray = None):
        ub = Wini[S].min()
        return ub - bini

    def lagrangian_cond(self,):
        if self.lambdat is not None and (self.lambdat[self.T] < -self.epsilon).any():
            return False
        
        if self.lambda0 is not None and self.lambda0 < -self.epsilon:
            return False

        return True


    def get_summary(self, Wini, bini, S, T=None):
        return {
            "b": self.b,
            "lambda0": self.lambda0,
            "loss": self.loss,
            "lambdat": self.lambdat[T] if self.lambdat is not None else None,
            "W": self.W,
            "left_cond_init": self.left_cond(Wini=Wini, bini=bini, S=S, epsilon=self.epsilon),
            "right_cond_init": self.right_cond(Wini=Wini, bini=bini, S=S),
            "left_cond_proj": self.left_cond(Wini=self.W, bini=self.b, S=S, epsilon=self.epsilon),
            "right_cond_proj": self.right_cond(Wini=self.W, bini=self.b, S=S),
            "left_margin_init": self.left_margin(Wini=Wini, bini=bini, S=S),
            "right_margin_init": self.right_margin(Wini=Wini, bini=bini, S=S),
            "left_margin_proj": self.left_margin(Wini=self.W, bini=self.b, S=S),
            "right_margin_proj": self.right_margin(Wini=self.W, bini=self.b, S=S),
            "lagrangian_cond": self.lagrangian_cond(),
        }

    def projection_left_right(self,):
        Wini, bini, T, S = self.Wini, self.bini, self.T, self.S
        omega = np.ones_like(Wini).astype(bool)

        # card_sbar = np.float32(omega.sum() - S.sum())
        card_sbar = (~S).sum()
        card_T = T.sum()
        wsum_T = Wini[T].sum()

        denom = np.float32(1 / (1 + card_sbar * (card_T + 1)))

        self.b = denom * (Wini[~S].sum() + card_sbar * (wsum_T + bini))
        self.lambda0 = denom * ((card_T + 1) * Wini[~S].sum() - wsum_T - bini)

        self.lambdat = np.zeros_like(Wini)
        self.lambdat[T] = self.b - Wini[T]

        W = np.zeros_like(Wini)
        W[T] = self.b
        W[S & ~T] = Wini[S & ~T]
        W[~S] = Wini[~S] - self.lambda0
        self.W = W

        # self.loss = ((Wini[T] - self.b)**2).sum() + card_sbar * self.lambda0 ** 2 + (self.b - bini) ** 2
        self.loss = 1/2 * (((self.W - Wini) ** 2).sum() + (self.b - bini) ** 2)

        self.summary = self.get_summary(Wini=Wini, bini=bini, S=S, T=T)

        return self

    def projection_left(self,):
        Wini, bini, T, S = self.Wini, self.bini, self.T, self.S
        omega = np.ones_like(Wini).astype(bool)
        card_sbar = omega.sum() - S.sum()
        
        self.b = ( 1 / (1 + card_sbar)) * (Wini[~S].sum() + card_sbar * bini)
        self.lambda0 = self.b - bini

        W = np.zeros_like(Wini)
        W[~S] = Wini[~S] - self.lambda0
        W[S] = Wini[S]
        self.W = W

        self.loss = 1/2 * (((self.W - Wini) ** 2).sum() + (self.b - bini) ** 2)

        self.summary = self.get_summary(Wini=Wini, bini=bini, S=S, T=T)

        return self

    def projection_right(self,):
        Wini, bini, T, S = self.Wini, self.bini, self.T, self.S
        self.b = (1 / (1 + T.sum())) * (Wini[T].sum() + bini)
        
        self.lambdat = np.zeros_like(Wini)
        self.lambdat[T] = self.b - Wini[T]

        W = np.zeros_like(Wini)
        W[T] = self.b
        W[~T] = Wini[~T]
        self.W = W

        self.loss = 1/2 * (((self.W - Wini) ** 2).sum() + (self.b - bini) ** 2)

        self.summary = self.get_summary(Wini=Wini, bini=bini, S=S, T=T)

        return self
```

```python
class MultiPlot:

    def __init__(self, nrow, ncol, *args, **kwargs):
        self.nrow = nrow
        self.ncol = ncol
        self.fig = plt.figure(*args, **kwargs)
    
    def add_subplot(self, pos, nrow=None, ncol=None, *args, **kwargs):
        if nrow is None:
            nrow = self.nrow
        if ncol is None:
            ncol = self.ncol
        if isinstance(pos, tuple):
            pos = (pos[0] - 1) * self.ncol + pos[1]
        return self.fig.add_subplot(nrow, ncol, pos, *args, **kwargs)
```
